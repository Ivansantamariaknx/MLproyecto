```python
# Definir una función personalizada para asignar el tipo de obra según el texto de las observaciones
def asignar_tipo_obra(texto):
    if 'general' in texto or 'integral' in texto or  'INTEGRAL' in texto  or  'GENERAL' in texto:
        return '1'
    elif 'BAÑO' in texto and 'COCINA' in texto or 'baño' in texto and 'cocina' in texto :
        return '2'
    elif 'baño' in texto or 'BAÑO' in texto:
        return '3'
    elif 'cocina' in texto or 'COCINA' in texto:
        return '4'
    else:
        return '5'

# Aplicar la función a la columna 'observaciones' y crear una nueva columna 'tipodeobra'
datosweb['tipodeobra'] = datosweb['OBSERVACIONES'].apply(asignar_tipo_obra)
```

# Oversampling

```python
from imblearn.over_sampling import RandomOverSampler

# Crear el objeto RandomOverSampler
oversampler = RandomOverSampler()

# Aplicar oversampling a los datos
X_oversampled, y_oversampled = oversampler.fit_resample(datoswebnums, datoswebnums["Resultado"])

# Crear un nuevo DataFrame con los datos oversampleados
datosweb_oversampled = pd.DataFrame(X_oversampled, columns=datoswebnums.columns)
datosweb_oversampled["Resultado"] = y_oversampled
```

```python
datosweb_oversampled ["Resultado"].value_counts()
```

```python
datosweb_oversampled.dtypes
```

# Defino x e y

```python
X = np.array(datosweb_oversampled.drop(['Resultado'], 1))
y = np.array(datosweb_oversampled['Resultado'])
```

```python
from sklearn import model_selection

validation_size = 0.20
seed = 7
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X,
                                                                    y,
                                                                    test_size=validation_size,
                                                                    random_state=seed)
```

```python
from sklearn.model_selection import cross_val_score
```

# Logistic regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(X_train, Y_train)
```

```python
prediction= model.predict(X_test)
```

```python
accuracyregression=  model.score(X_test, Y_test)
accuracyregression
```

```python
name='Logistic Regression'
kfold = model_selection.KFold(n_splits=10) #Parte los datos en 10 trozos para usar validación cruzada / cross validation
cv_results = model_selection.cross_val_score(model, X_train, Y_train, cv=kfold, scoring='accuracy')

msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
print(cv_results)
```

# Feature Importance

```python
from sklearn.ensemble import BaggingRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
bagging_model = BaggingRegressor(base_estimator=DecisionTreeRegressor())
```

```python
bagging_model.fit(X_train, Y_train)
```

```python
#feature importance


feature_importances = []
for base_model in bagging_model.estimators_:
    feature_importances.append(base_model.feature_importances_)
```

```python
average_importance = np.mean(feature_importances, axis=0)
```

```python
sorted_indices = np.argsort(average_importance)[::-1]
```

```python
X_train_df = pd.DataFrame(X_train, columns=['MES', 'COMERCIAL_NUM', 'TELEOP_NUM', 'CANAL_NUM', 'tipodeobra'])
```

```python
sorted_features = X_train_df.columns[sorted_indices]
```

```python
for feature, importance in zip(sorted_features, average_importance):
    print(f"{feature}: {importance}")
```
